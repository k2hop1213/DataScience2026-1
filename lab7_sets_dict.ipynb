## Sets
- unordered collection of distinct objects

```python
[] # empty list
type((10,))# 1 element tuple
x = set()
type(x)
print(x)
```

```python
low = {0,1,2,3,4,2}
print(low)
odds = [1,3,5,7,9,5]
odds = set(odds)

odds.remove(5)
odds.add(11)
print(odds)
```

```python
# low.intersection(odds)
# low.union(odds)
# low.difference(odds)
low&odds
low|odds
low - odds

ten = set(range(10))

# low.issubset(ten)

low <= ten
low < ten

low.clear()
print(low)
```

```python
myset = {1,2,'hello',(1,2,3),[2,3]} #바꿀 수 있는건 못들어감 unhashable type

for i in myset:
    print(i)
```

```python
# hashable type
x = (1,2,4)
y = (1,2,3)
print(id(x),id(y))
print(x is y)
print(x == y)
print(x.__hash__(),y.__hash__()) #내용이 같으면 해쉬값이 같음
#내용이 바뀌면 해쉬값도 바뀜
#mutable한 오브젝트는 unhashable
```

## dictionary
- key와 value의 쌍으로 되어있음
- **ordered** mutable collection

```python
{} #empty dictionary
```

```python
birthday = {'Newton':1642, 'Darwin':1809} # key: value, key는 hashable한 object만
birthday['Darwin'] = 1808
birthday['Turing'] = 1912
# del birthday['Newton']
print(birthday)
```

```python
'Turing' in birthday
'Newton' in birthday
```

```python
# for x in birthday:
#     print(x, birthday[x])

for k,v in birthday.items():
    print(k,v)
```

```python
# histogram
birds = ['tern','goose','goose','hawk','tern','goose','tern']
observations ={}
# for i in birds:
#     if i in observations:
#         observations[i] += 1
#     else :
#         observations[i] = 1

for i in birds:
    observations[i] = observations.get(i,0) + 1

print(observations)
print(observations.keys())
print(observations.values())
print(observations.items())
```
