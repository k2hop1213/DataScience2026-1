```python
x = 4

if x % 2: #0, None, False
    print(x,'is odd')
else :
    print(x,'is even')
```

```python
a = int(input('what is your height (cm)?'))
b = int(input('what is your weight (Kg)?'))
a = a/100
bmi = b/(a*a) # **2

if bmi < 18:
    print('underweight')
elif bmi<25:
    print('normal weight')
elif bmi<27:
    print('overweight')
elif bmi<30:
    print('moderate obese')
elif bmi<40:
    print('severe obese')
else :
    print('very severe obese') 
```
