```python
#1

class Classroom:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        if isinstance(y,str) :
            self.y = [y]

            
    def __str__(self):
        return f"{self.x}_Grade : {self.y}"
    
    def add_student(self, grade, name):
        if self.x != grade :
            return "Invalid Input"
        else :
            if isinstance(name,str) :
                self.y += [name]
            else :
                self.y += name

    def __len__ (self):
        return len(self.y)
    
f_grade = Classroom(1,"sunwoo")

print(f_grade)
```

```python
#2
f_grade.add_student(1,"Kim")
print(f_grade)
```

```python
#3
print(len(f_grade))
```

```python
#4

info = {
    "high_school":{
        "1":["sunwoo","Lee"],
        "2":"Kim",
        "3":"park"
    },
    "middle_school":{
        "2":["Andrew","Yoshua"]
    }
}

school_data = {}

for i,j in info.items():
    school_data[i] = []
    
    for k,l in j.items():

        a = Classroom(k, l)
        school_data[i].append(a)

for i in school_data:
    print(i)
    for j in school_data[i]:
        print(j)
          
```

```python
#5

high_school_num = [{f"{c.x}grade": len(c)} for c in school_data["high_school"]]
middle_school_num = [{f"{c.x}grade": len(c)} for c in school_data["middle_school"]]


print(high_school_num)
print(middle_school_num)
```

```python
#6

h_sum = sum([list(d.values())[0] for d in high_school_num])
m_sum = sum([list(d.values())[0] for d in middle_school_num])

print(f"high_school : {h_sum}")
print(f"middle_school : {m_sum}")
```
