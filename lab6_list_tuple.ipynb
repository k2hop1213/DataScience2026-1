```python
fruits = ["apple","kiwi","banana","grape","mango"]

for idx,fruit in enumerate(fruits):
    print(f"{idx} : {fruit}")
```

```python
#List aliases
a = [1,2,3]
b=a
b[0]=100
print(a) #이름이 여러가지가 되.
#b=a[:] #copy
b = a.copy()
b[0]=200
print(a)
```

```python
import copy

a = [1,[2,3]]
#b = a[:]
b = copy.deepcopy(a) #완전히 분리
b[1][0] = 300
print(a)
```

```python
def double(mylist):
    for i,v in enumerate(mylist):
        mylist[i] = 2*v


things = [1,2,'hello']
double(things)
print(things)
```

```python
def double(seq): #int 는 immutable
    seq = seq*2


things = 10
double(things)
print(things)
```

```python
li =[('a',1),('c',3),('b',2)]
[n for k,n in li]
```

```python
num_string ='1 2 3 4 5 6'

num_string2 = num_string.split(' ')
num_string2 =[int(i) for i in num_string2]

num_pairs = [(1,2),(3,4),(5,6)]
num_pairs_revers = [(s,f) for f,s in num_pairs]

print(num_string2)
print(num_pairs_revers)
```
