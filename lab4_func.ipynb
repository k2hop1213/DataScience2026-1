```python
#defining a function

def double(x: int)-> int:#파라미터
    #입력값의 타입을 강제할 수 없다.
    #힌트만 줄뿐 코드에 에러가 나지 않는다.
    """
    Calculate a doulbe of x
    Args: x: int input
    returns: twice of x
    """
    return 2*x
```

```python
double(30)#argument
double('hello')
double([1,2,3,'hello'])
```

```python
def print_square(n: int) :

    for i in range(n):
        for j in range(n):
            print("#",end='')
        print()

#함수도 object이다. 코드가 들어가있는 object,,
print_square(3)
```

```python
#scope 내가 볼 수 있는 변수의 영역
#재귀호출
#recursive한 구조와 탈출조건이 필요함
def factorial(x):
    if x == 1:
        return 1
    
    return x*factorial(x-1)
```

```python
factorial(5)
```

```python
def sum_to_num(n):
    if n == 1:
        return 1
    return n+sum_to_num(n-1)

sum_to_num(100)
```
