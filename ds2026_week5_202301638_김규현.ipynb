```python
#6
def Roma(num):
    Roman=["I","II","III","IV","V","VI","VII","VIII","IX","X"]
    return Roman[num-1]

n = int(input("1 ~ 10 사이의 숫자를 입력하세요:"))
print(Roma(n))
```

```python
#7
R_num=[Roma(i) for i in range(1,11)]
print(R_num)
```

```python
#8
for n,r in enumerate(R_num):
    print(f"{n + 1} : {r}")
```
