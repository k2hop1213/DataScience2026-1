```python
#8
with open('the-blue-hotel.txt','r',encoding='utf-8') as reader:
    a = reader.readline()
    res = 0
    while a:
        res += len(a.split())
        a=reader.readline()
    print(res)
```

```python
#9
with open('the-blue-hotel.txt','r',encoding='utf-8') as reader:
    a = reader.readline()
    b,s,d = 0, 0, 0
    while a:
        b += a.count('b') + a.count('B')
        d += a.count('d') + a.count('D')
        s += a.count('s') + a.count('S')
        a=reader.readline()
    print(f'B word uses : {b}')
    print(f'S word uses : {s}')
    print(f'D word uses : {d}')
```

```python
#10
with open('Atoz.txt','w',encoding='utf-8') as writer:
    a = ''
    for i in range(26):
        a += chr(ord('a') + i)
    writer.write(a)
```

```python
#11
with open('Atoz.txt','r',encoding='utf-8') as reader:
    a = reader.read()
    b = {}
    for i in a:
        b[i] = 0

print(b)
```

```python
#12
with open('the-blue-hotel.txt', 'r', encoding='utf-8') as reader:
    c = reader.read()
    for i in c:
        char = i.lower()
        if char in b: 
            b[char] += 1

print(b)
```

```python
#13
print(f'0th minimum alphabet {min(b, key = b.get)} : {min(b.values())}')
print(f'0th maximum alphabet {max(b, key = b.get)} : {max(b.values())}')
#13-1
b1 = sorted(b.items(), key=lambda item: item[1])
for i in range(3):
    print(f'{i}th minimum alphabet = {b1[i][0]} : {b1[i][1]}')
    print(f'{i}th maximum alphabet = {b1[-(i + 1)][0]} : {b1[-(i + 1)][1]}')
```
