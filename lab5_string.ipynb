## String
- an ordered sequence of smaller parts: characters
- a character is not a distinct data type, but but a string of length 1
- use bracket operator for indexing:[ ]

```python
city = 'incheon'
type(city)
city[4]
```

```python
dir(city)
```

```python
city = 'incheon'
len(city)
mylen = len(city)
city[mylen] #error
# 음수로 인덱스 접근하면 뒤에서부터 읽음
```

```python
#characters
ord('A')
chr(65)
for i in range(10):
    print(chr(0x1F642 + i))
```

```python
a = ord('a')
for i in range(26):
    print(chr(a + i),end=' ')
```

```python
#모음 제거
s = input()
print(s)
for i in s:
    if i not in 'aeiou':
        print(i,end='')
    else :
        print(' ',end='')
        
```
