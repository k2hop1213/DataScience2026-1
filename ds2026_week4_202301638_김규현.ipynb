```python
#3
def check_UTC(x):
    a = time.gmtime()
    if x == 'year':
        print(a.tm_year)
    elif x == 'month':
        print(a.tm_mon)
    elif x == 'day':
        print(a.tm_mday)
    elif x == 'hour':
        print(a.tm_hour)
    elif x =='min':
        print(a.tm_min)
    elif x =='sec':
       print(a.tm_sec)
    else :
        print("wrong input! please try different input!")
        return
    print(x)

time_part = input("Enter want to see time part: ")
check_UTC(time_part)
```

```python
#4
def recursion_power(n):
    """Sum 1 to n then, each number is squared!
    Args :
        n:
         input is int and destination number
    Returns :
        Output is summation 1^2 to n^2
    """
    if n == 1:
        return 1

    return (n*n) + recursion_power(n-1)

print(recursion_power(4))
```

```python
#5
def loop_power(n):
    res = 0
    for i in range(1,n+1):
        res += (i*i)
    return res

print(loop_power(5))
```

```python
#7

lps = loop_power
print(lps(5))
'''
lps가 loop_power를 가리키게 되어서
lps는 이제 loop_power를 나타내는 또 다른 이름이 된것입니다
그래서 loop_power와 동일하게 사용할 수 있게 된거 같습니다
'''
lps.__name__

'''
lps가 loop_power를 가리키기 때문에 함수를 사용할때와 동일한 method를 사용할 수 있게되어
__name__을 출력했을때 현재 가리키고 있는 함수의 이름이 나오는거 같습니다
'''
```

```python
#8
import time

#8-1
power_sum = [recursion_power,loop_power]

#8-2
def check_func_time(func,num):
    start_point = time.perf_counter_ns()
    result = func(num)
    end_point = time.perf_counter_ns()
    return [result, end_point - start_point]
result_time = 0.0
iter = 5
for i in power_sum:
    res = check_func_time(i,iter)
    for j in range(1,iter + 1):
        print(res[0])
    print(f"{i.__name__}: {res[1]}")
```

```python
#10

def unfrog(x):
    if type(x) == int:
        print(chr(x))
    else :
        print(ord(x))

unfrog(77)
```

```python
#11
def ununfrog(x):
    if type(x) == int:
        print(chr(x))
    else :
        print(bin(ord(x)))
        print(oct(ord(x)))
        print(hex(ord(x)))

ununfrog('B')
```
