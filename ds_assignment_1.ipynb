```python

print("--- Case #1 ---")
ax1, ay1, ax2, ay2 = 1, 2, 4, 5
bx1, by1, bx2, by2 = 2, 1, 5, 4

ix1 = max(ax1, bx1)
iy1 = max(ay1, by1)
ix2 = min(ax2, bx2)
iy2 = min(ay2, by2)

width = ix2 - ix1
height = iy2 - iy1

if width > 0 and height > 0:
    area = width * height
    print(f"overlap!")
    print(f"area: {area}")
else:
    print(f"no overlap")
    print(f"area: 0")
```
