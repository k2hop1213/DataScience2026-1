```python
#read in chunks
reader = open('alice.txt','r',encoding='utf-8')
data = reader.read(64)
while data:
    print(len(data))
    data = reader.read(64)
```

```python
#read in chunks
reader = open('alice.txt','r',encoding='utf-8')
data = reader.readline()
while data:
    print(len(data))
    data = reader.readline()
```

```python
reader = open('alice.txt','r',encoding='utf-8')
data = reader.readline()
num_chars , num_words, num_lines =0,0,0
while data:
    num_lines += 1
    num_chars += len(data)
    num_words += len(data.split())

    data = reader.readline()

print(num_chars,num_words,num_lines)
print(f"{num_chars/num_lines:.2f}")
print(f"{num_words/num_lines:.2f}")
```

```python
reader = open('alice.txt','r',encoding='utf-8')

num_chars , num_words, num_lines =0,0,0
for line in reader:
    num_chars+=len(line)
    num_words+=len(line.split())
    num_lines+=1
reader.close()
print(num_chars,num_words,num_lines)
print(f"{num_chars/num_lines:.2f}")
print(f"{num_words/num_lines:.2f}")
```
