## loop(반복문)

```python
n=10
while n>0: #True 무한루프, False 건너뛰기
    print(n)
    n = n-1
print('Blast off!')
```

```python
n=1
while n<=10:
    print(n,n**2,n**3,sep='\t') #default는 스페이스로 구분
    n+=1
```

```python
n = 1
while n<= 10:
    i = 1
    while i<=10:
        print(n*i,end = '\t')
        i+=1
    print('\n')
    n+=1
```
